```python
dimensionality = mean_duration * 44000 * 2
print('sample dimensionality: {:.1e}'.format(dimensionality))
print('total size, i.e. number of audio samples: {:.1e}'.format(dimensionality * len(tracks)))
```

Dimensionality refers to the total number of individual audio samples required to represent one track of average duration. It is a measure of the size of the data for a single track in terms of raw audio samples.

Breaking It Down
Audio Representation:

1. Audio is represented digitally as a sequence of samples, where each sample is a numerical value representing the sound's amplitude at a specific time.
The sampling rate determines how many samples are taken per second. A common sampling rate is 44,000 Hz (samples/second).
2. Formula for Dimensionality:

dimensionality = mean_duration * 44000 * 2

* mean_duration: Average track duration in seconds.
* 44000: Sampling rate (44,000 samples per second).
* 2: Likely refers to stereo sound (2 channels: left and right).

So, dimensionality is the total number of audio samples for a single track of average duration.

### 1.1 Splits

```python
SPLITS = ['training', 'validation', 'test']
SUBSETS = ['small', 'medium', 'large']
print('subset    #train    #val   #test  val_ratio test_ratio')
for subset in SUBSETS:
    counts = [sum((tracks['set', 'split'] == split) & (tracks['set', 'subset'] <= subset)) for split in SPLITS]
    #ratios = np.array(counts[0] / counts[1:])
    ratios = counts[0] / np.array(counts[1:])
    print('{:8s} {:7d} {:7d} {:7d} {:8.2f} {:9.2f}'.format(subset, *counts, *ratios))
```

```python
for subset in ['small', 'medium']:
    subset = tracks['set', 'subset'] <= subset

    d = genres.reset_index().set_index('title')
    d = d.loc[tracks.loc[subset, ('track', 'genre_top')].unique()]

    for split in SPLITS:
        b = tracks['set', 'split'] == split
        d['#' + split] = tracks.loc[subset & b, ('track', 'genre_top')].value_counts()

    d['val_ratio'] = d['#training'] / d['#validation']
    d['test_ratio'] = d['#training'] / d['#test']

    ipd.display(d.sort_values('#training', ascending=False))
```

```python
d = pd.DataFrame(index=genres.index, columns=SPLITS)
for genre in genres.index:
    b = tracks['track', 'genres_all'].map(lambda genres: genre in genres)
    d.loc[genre] = tracks.loc[b, ('set', 'split')].value_counts()
d['val_ratio'] = d['training'] / d['validation']
d['test_ratio'] = d['training'] / d['test']
d.sort_values('training', ascending=False, inplace=True)
ipd.display(d.head(10))
ipd.display(d.tail(10))
```

```python


def isnull(column, df):
    """Check if the column has null or invalid values."""
    if column[1] in ['tags', 'genres', 'genres_all']:
        # Check if lists in the column are empty
        return df[column].apply(lambda x: len(x) == 0 if isinstance(x, (list, tuple)) else True)
    elif pd.api.types.is_numeric_dtype(df[column]):  # Handle numeric columns
        return df[column] <= 0
    else:
        # Check for actual null values in other data types
        return df[column].isnull()

def count(column_name):
    """Count non-null values and calculate percentage for a given column."""
    col0 = column_name[0]
    df = tracks if col0 == 'track' else tracks.drop_duplicates(subset=(col0, 'id'))
    # Correctly identify non-null entries
    non_null_entries = ~isnull(column_name, df)
    n = non_null_entries.sum()  # Count valid values
    p = n / len(df) * 100  # Calculate percentage
    return n, p

# Prepare DataFrame to store results
columns_to_evaluate = tracks.columns.drop('set')  # Exclude 'set' column
results = [count(col) for col in columns_to_evaluate]  # Get counts for each column

# Create DataFrame from results
d = pd.DataFrame(results, index=columns_to_evaluate, columns=['n', 'p'])
d['n'] = d['n'].astype(int)  # Convert 'n' to integer
d
```

### 4.2 Cross-appearance

Todo:

```python
enc = MultiLabelBinarizer()
genres_indicator = enc.fit_transform(tracks['track', 'genres'])
genres_names = enc.classes_
genres_names = genres.loc[enc.classes_, 'title'].values
cross_correlation = genres_indicator.T @ genres_indicator
```

```python
np.fill_diagonal(cross_correlation, 0)

plt.figure(figsize=(28, 28))
plt.imshow(np.log(cross_correlation))
plt.yticks(range(len(genres_names)), genres_names);
plt.xticks(range(len(genres_names)), genres_names, rotation=90);
```

```python
cross_correlation = np.tril(cross_correlation, k=-1)
sort = np.argsort(cross_correlation.flatten())

N = 20
indices = np.unravel_index(sort[:-N:-1], cross_correlation.shape)
for i, j in zip(*indices):
    print('{}: {} | {}'.format(cross_correlation[i, j], genres_names[i], genres_names[j]))
```
